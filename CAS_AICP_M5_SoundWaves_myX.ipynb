**Basics of making Sounds**

Code sniped to create sound (how to make 1-2 tones)

Tips: Google or Gemini "what is the Audio Library of Python"


```python
#create two waves and play with them
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile
from IPython.display import Audio

f1=440
f2=450
fs=44100 #sampling requency

T=3 #3 seconds long sound
twopi=2*np.pi

t=np.linspace(0,T,T*fs)

output=np.sin(twopi*f1*t)+np.sin(twopi*f2*t)

#plt.plot(t,output)
#plt.show()
Audio(output,rate=fs)
```
